Ce notebook reproduit les analyses faites par `scripts/run_eda.py`. Il affiche les graphiques générés et charge le CSV récapitulatif.

Les images et CSV sont dans le dossier `docs/`. Si vous voulez régénérer les artefacts, exécutez `python scripts/run_eda.py`.

# Imports et configuration
import os
import pandas as pd
from IPython.display import Image, display
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

DOCS = os.path.join('..', 'docs')
GRAPHS = os.path.join(DOCS, 'graphs')

print('✅ Configuration OK')

# Afficher le résumé des stats exporté par scripts/run_eda.py
stats_path = os.path.join(DOCS, 'eda_stats_recap.csv')
if os.path.exists(stats_path):
    stats = pd.read_csv(stats_path)
    display(stats.head())
else:
    print(f'Le fichier récapitulatif nexiste pas : {stats_path}')

# Afficher quelques graphiques générés (si présents)
example_graphs = [
    '01_distribution_consommation_kwh.png',
    '02_correlation_recharge_montant.png',
    '03_evolution_post_2026.png',
    '04_economies_tranche_sociale.png',
    '05_surcouts_t2_t3.png',
    '06_synthese_eda.png'
]
for g in example_graphs:
    p = os.path.join(GRAPHS, g)
    if os.path.exists(p):
        display(Image(filename=p))
    else:
        print(f'Manquant: {p}')

# Charger rapidement les jeux de données sources et afficher les tailles
raw_base = os.path.join('..', 'data', 'raw')
paths = {
    'zones': os.path.join(raw_base, 'zones_senegal.csv'),
    'users': os.path.join(raw_base, 'users.csv'),
    'consumption': os.path.join(raw_base, 'consumption_daily.csv'),
    'recharges': os.path.join(raw_base, 'recharges.csv'),
}
datasets = {}
for name, p in paths.items():
    if os.path.exists(p):
        if name in ['consumption', 'recharges']:
            datasets[name] = pd.read_csv(p, parse_dates=['date'])
        else:
            datasets[name] = pd.read_csv(p)
        print(f'✅ {name:12s}: {len(datasets[name]):,} lignes')
    else:
        print(f'Fichier source manquant: {p}')
